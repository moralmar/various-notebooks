Let's try that spiral function now in Python

```python
import math
import numpy as np
import pandas as pd

def two_spiral_inPy(N = 200, radians = 3*math.pi, theta0 = math.pi/2, labels = [0,1]):

    N1 = math.floor(N/2)
    N2 = N - N1

    theta = theta0 + np.random.uniform(0, 1, N1) * radians
    
    # note 1: math.cos doesn't play well with other np.functions 
    # note 2: pd.concat takes only(?) pd.DataFrames
    
    spiral_11 = -theta * np.cos(theta) + np.random.uniform(0, 1, N1)
    spiral_11 = pd.DataFrame(spiral_11)
    spiral_12 = theta * np.sin(theta) + np.random.uniform(0,1, N1)
    spiral_12 = pd.DataFrame(spiral_12)
    
    spiral_1 = pd.concat([spiral_11, spiral_12], axis = 1) # output: 100x2 DataFrame
    
    
    spiral_21 = theta * np.cos(theta) + np.random.uniform(0, 1, N2)
    spiral_21 = pd.DataFrame(spiral_21)
    spiral_22 = -theta * np.sin(theta) + np.random.uniform(0, 1, N2)
    spiral_22 = pd.DataFrame(spiral_22)
    
    spiral_2 = pd.concat([spiral_21, spiral_22], axis = 1)
    
    points  = spiral_1.append(spiral_2).reset_index(drop=True) # output: 200x2
    classes = np.repeat(labels, [N1,N2], axis = 0) 
    classes = pd.Categorical(classes, categories=labels)
    dat = pd.concat([points, pd.Series(classes)], axis = 1)
    
    dat.columns = ['x1', 'x2', 'class']
    
    
    return dat

data = two_spiral_inPy(labels=['non hot dog','hot dog'])
data.head()
```
